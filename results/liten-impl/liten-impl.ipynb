```python
import os
import torch
import random
import numpy as np
import json
import time
from PIL import Image
from dataclasses import dataclass, field
from typing import List, Dict, Optional, Tuple
import imageio

# Configuration
@dataclass
class Config:
    SEED: int = 2048
    DEVICE: str = "cuda" if torch.cuda.is_available() else "cpu"

    # Checkpoints
    CHECKPOINT_SPATIAL: str = "openvla/openvla-7b-finetuned-libero-spatial"
    CHECKPOINT_OBJECT: str = "openvla/openvla-7b-finetuned-libero-object"

    # LITEN parameters
    MAX_RETRIES: int = 3  

    # Environment
    NUM_STEPS_WAIT: int = 10
    NUM_TRIALS_PER_TASK: int = 5
    NUM_TASKS_TO_RUN: int = 5
    RESOLUTION: int = 256
    RESIZE_SIZE: int = 224
    CENTER_CROP: bool = True

    # Task-specific
    MAX_STEPS_SPATIAL: int = 220
    MAX_STEPS_OBJECT: int = 280

    # Logging
    LOG_DIR: str = "./results/logs"
    VIDEO_DIR: str = "./results/rollouts"
    SAVE_VIDEOS: bool = True

config = Config()

# Set seed for reproducibility
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True

set_seed(config.SEED)

print(f"Configuration loaded")
print(f"  Device: {config.DEVICE}")
print(f"  LITEN Max Retries: {config.MAX_RETRIES}")

# ============================================================================
# Affordance Reasoning System
# ============================================================================

@dataclass
class ExecutionExperience:
    """Records a single execution attempt for learning."""
    task_description: str
    affordance_hints: List[str]
    success: bool
    num_steps: int
    failure_mode: Optional[str] = None

class LITENReasoner:
    """
    LITEN's affordance reasoning component.

    Key Innovation: Instead of just running the VLA once, LITEN:
    1. Generates affordance reasoning before execution
    2. Observes execution outcome
    3. Updates reasoning based on success/failure
    4. Retries with refined understanding

    This PoC uses heuristic reasoning (no external VLM needed).
    """

    def __init__(self):
        self.experience_buffer: List[ExecutionExperience] = []
        self.task_insights: Dict[str, List[str]] = {}

    def generate_affordance_hints(
        self,
        task_description: str,
        attempt_number: int = 1,
        past_failure_mode: str = None
    ) -> List[str]:
        """
        Generate affordance hints for the task.

        Returns hints about:
        - Object affordance: what to manipulate
        - Grasp affordance: how to grasp
        - Spatial affordance: where to place
        - Movement affordance: how to move
        """
        hints = []
        task_lower = task_description.lower()

        # === OBJECT AFFORDANCE ===
        if "black bowl" in task_lower:
            hints.append("TARGET: black bowl (round, dark container)")
            # Spatial context from task
            if "between" in task_lower:
                hints.append("LOCATION: bowl is between two objects")
            elif "next to" in task_lower:
                hints.append("LOCATION: bowl is adjacent to another object")
            elif "table center" in task_lower:
                hints.append("LOCATION: bowl at table center")
            elif "cookie box" in task_lower:
                hints.append("LOCATION: bowl on top of cookie box")
            elif "drawer" in task_lower:
                hints.append("LOCATION: bowl inside drawer")
        elif "soup" in task_lower:
            hints.append("TARGET: soup can (cylindrical container)")
        elif "cream cheese" in task_lower:
            hints.append("TARGET: cream cheese box (rectangular)")
        elif "ketchup" in task_lower:
            hints.append("TARGET: ketchup bottle (red, bottle shape)")
        elif "salad dressing" in task_lower:
            hints.append("TARGET: salad dressing bottle")
        elif "bbq sauce" in task_lower:
            hints.append("TARGET: BBQ sauce bottle")

        # === GRASP AFFORDANCE ===
        if "bowl" in task_lower:
            hints.append("GRASP: approach from above, grip bowl rim")
        elif any(x in task_lower for x in ["sauce", "dressing", "ketchup"]):
            hints.append("GRASP: grip bottle body at middle")
        else:
            hints.append("GRASP: stable grip on accessible surface")

        # === SPATIAL AFFORDANCE ===
        if "plate" in task_lower:
            hints.append("PLACE: center of plate (flat surface)")
        elif "basket" in task_lower:
            hints.append("PLACE: inside basket (open container)")

        # === MOVEMENT AFFORDANCE ===
        if "between" in task_lower:
            hints.append("MOVE: careful navigation between objects")
        elif "drawer" in task_lower:
            hints.append("MOVE: precise movement in confined space")
        else:
            hints.append("MOVE: direct path to target")

        # === LEARNING FROM PAST ATTEMPTS ===
        if attempt_number > 1 and past_failure_mode:
            if "grasp" in past_failure_mode.lower():
                hints.append("RETRY: adjust grasp position/angle")
            elif "place" in past_failure_mode.lower():
                hints.append("RETRY: adjust placement height/position")
            elif "timeout" in past_failure_mode.lower():
                hints.append("RETRY: use more direct movement")
            else:
                hints.append("RETRY: try alternative approach")

        # Add task-specific learned insights
        if task_description in self.task_insights:
            for insight in self.task_insights[task_description][-2:]:
                hints.append(f"LEARNED: {insight}")

        return hints

    def assess_execution(
        self,
        task_description: str,
        success: bool,
        num_steps: int,
        max_steps: int
    ) -> str:
        """
        Assess what happened during execution.
        Returns a failure mode description if failed.
        """
        if success:
            # Learn from success
            insight = f"Success pattern: completed in {num_steps} steps"
            if task_description not in self.task_insights:
                self.task_insights[task_description] = []
            self.task_insights[task_description].append(insight)
            return "SUCCESS"

        # Analyze failure mode based on when it failed
        progress = num_steps / max_steps

        if progress < 0.2:
            failure_mode = "EARLY_FAILURE: likely grasp failure or wrong target"
        elif progress < 0.5:
            failure_mode = "MID_FAILURE: object manipulation issue"
        elif progress < 0.8:
            failure_mode = "LATE_FAILURE: placement or navigation issue"
        else:
            failure_mode = "TIMEOUT: nearly completed, minor adjustment needed"

        # Store insight for future
        if task_description not in self.task_insights:
            self.task_insights[task_description] = []
        self.task_insights[task_description].append(failure_mode)

        return failure_mode

    def record_experience(self, experience: ExecutionExperience):
        """Store experience for in-context learning."""
        self.experience_buffer.append(experience)
        if len(self.experience_buffer) > 100:
            self.experience_buffer = self.experience_buffer[-100:]

# Initialize global reasoner
reasoner = LITENReasoner()
print("LITEN Reasoner initialized")

# ============================================================================
# Model Loading and Utility Functions
# ============================================================================

import tensorflow as tf
from prismatic.extern.hf.configuration_prismatic import OpenVLAConfig
from prismatic.extern.hf.modeling_prismatic import OpenVLAForActionPrediction
from transformers import (
    AutoConfig, AutoImageProcessor, AutoModelForVision2Seq, AutoProcessor
)
from prismatic.extern.hf.processing_prismatic import (
    PrismaticImageProcessor, PrismaticProcessor
)
from libero.libero import benchmark, get_libero_path
from libero.libero.envs import OffScreenRenderEnv
import math

# Register OpenVLA
AutoConfig.register("openvla", OpenVLAConfig)
AutoImageProcessor.register(OpenVLAConfig, PrismaticImageProcessor)
AutoProcessor.register(OpenVLAConfig, PrismaticProcessor)
AutoModelForVision2Seq.register(OpenVLAConfig, OpenVLAForActionPrediction)

def load_model(checkpoint_path):
    """Load OpenVLA model."""
    model = AutoModelForVision2Seq.from_pretrained(
        checkpoint_path,
        attn_implementation="flash_attention_2",
        torch_dtype=torch.bfloat16,
        low_cpu_mem_usage=True,
        trust_remote_code=True,
    ).to(config.DEVICE)

    # Load normalization stats
    stats_path = os.path.join(checkpoint_path, "dataset_statistics.json")
    if os.path.isfile(stats_path):
        with open(stats_path, "r") as f:
            model.norm_stats = json.load(f)

    processor = AutoProcessor.from_pretrained(checkpoint_path, trust_remote_code=True)
    return model, processor

def resize_image(img, size):
    """Resize with JPEG encode/decode (matches training)."""
    img = tf.image.encode_jpeg(img)
    img = tf.io.decode_image(img, expand_animations=False, dtype=tf.uint8)
    img = tf.image.resize(img, size, method="lanczos3", antialias=True)
    return tf.cast(tf.clip_by_value(tf.round(img), 0, 255), tf.uint8).numpy()

def get_image(obs):
    """Get preprocessed image from observation."""
    img = obs["agentview_image"][::-1, ::-1]  # Rotate 180
    return resize_image(img, (config.RESIZE_SIZE, config.RESIZE_SIZE))

def center_crop(img_pil, crop_scale=0.9):
    """Apply center crop augmentation."""
    img = tf.convert_to_tensor(np.array(img_pil))
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.expand_dims(img, 0)

    s = float(np.sqrt(crop_scale))  # Use numpy instead of tf
    offset = (1 - s) / 2
    boxes = tf.constant([[offset, offset, offset + s, offset + s]], dtype=tf.float32)

    img = tf.image.crop_and_resize(img, boxes, [0], (224, 224))
    img = tf.clip_by_value(img[0], 0, 1)
    img = tf.image.convert_image_dtype(img, tf.uint8, saturate=True)
    return Image.fromarray(img.numpy()).convert("RGB")

def quat2axisangle(quat):
    """Convert quaternion to axis-angle."""
    if quat[3] > 1.0: quat[3] = 1.0
    elif quat[3] < -1.0: quat[3] = -1.0
    den = np.sqrt(1.0 - quat[3] * quat[3])
    if math.isclose(den, 0.0):
        return np.zeros(3)
    return (quat[:3] * 2.0 * math.acos(quat[3])) / den

def get_action(model, processor, img, task_description, unnorm_key):
    """Get action from OpenVLA."""
    pil_img = Image.fromarray(img).convert("RGB")
    if config.CENTER_CROP:
        pil_img = center_crop(pil_img)

    prompt = f"In: What action should the robot take to {task_description.lower()}?\nOut:"
    inputs = processor(prompt, pil_img).to(config.DEVICE, dtype=torch.bfloat16)

    action = model.predict_action(**inputs, unnorm_key=unnorm_key, do_sample=False)

    # Normalize gripper [0,1] -> [-1,1] and binarize
    action[..., -1] = np.sign(2 * action[..., -1] - 1)
    # Invert gripper
    action[..., -1] *= -1

    return action

print("Utility functions defined")

# ============================================================================
# Episode Runner
# ============================================================================

def run_episode_liten(
    env, model, processor, task_description, initial_state,
    max_steps, unnorm_key, attempt=1, past_failure=None
):
    """
    Run a single episode with LITEN enhancement.

    Key difference from baseline:
    - Generates affordance hints before execution
    - Tracks execution for assessment
    - Returns info for potential retry
    """
    # Reset environment
    env.reset()
    obs = env.set_init_state(initial_state)

    # LITEN: Generate affordance reasoning
    hints = reasoner.generate_affordance_hints(
        task_description,
        attempt_number=attempt,
        past_failure_mode=past_failure
    )

    # Log hints
    print(f"      Affordance hints: {hints[:2]}...")

    # Run episode
    t = 0
    replay_images = []
    done = False

    while t < max_steps + config.NUM_STEPS_WAIT:
        # Wait phase
        if t < config.NUM_STEPS_WAIT:
            obs, _, done, _ = env.step([0, 0, 0, 0, 0, 0, -1])
            t += 1
            continue

        # Get image and action
        img = get_image(obs)
        replay_images.append(img)
        action = get_action(model, processor, img, task_description, unnorm_key)

        # Execute
        obs, _, done, _ = env.step(action.tolist())
        if done:
            break
        t += 1

    num_steps = t - config.NUM_STEPS_WAIT

    # LITEN: Assess execution
    failure_mode = reasoner.assess_execution(
        task_description, done, num_steps, max_steps
    )

    # Record experience
    reasoner.record_experience(ExecutionExperience(
        task_description=task_description,
        affordance_hints=hints,
        success=done,
        num_steps=num_steps,
        failure_mode=failure_mode if not done else None
    ))

    return done, num_steps, replay_images, failure_mode

print("LITEN episode runner defined")

# ============================================================================
# Main Evaluation Loop with LITEN
# ============================================================================

def run_liten_evaluation(task_suite="SPATIAL"):
    """
    Run LITEN-enhanced evaluation.

    Key innovation: retry failed episodes with refined affordance reasoning.
    """
    # Setup
    if task_suite.upper() == "SPATIAL":
        checkpoint = config.CHECKPOINT_SPATIAL
        task_name = "libero_spatial"
        max_steps = config.MAX_STEPS_SPATIAL
    else:
        checkpoint = config.CHECKPOINT_OBJECT
        task_name = "libero_object"
        max_steps = config.MAX_STEPS_OBJECT

    print(f"\n{'='*80}")
    print(f"LITEN EVALUATION: {task_name}")
    print(f"Max Retries: {config.MAX_RETRIES}")
    print(f"{'='*80}\n")

    # Load model
    model, processor = load_model(checkpoint)
    print(f"Loaded model from {checkpoint}")

    # Initialize environment
    benchmark_dict = benchmark.get_benchmark_dict()
    task_suite_obj = benchmark_dict[task_name]()
    num_tasks = min(config.NUM_TASKS_TO_RUN, task_suite_obj.n_tasks)

    # Determine unnorm_key
    unnorm_key = task_name
    if hasattr(model, 'norm_stats') and unnorm_key not in model.norm_stats:
        if f"{unnorm_key}_no_noops" in model.norm_stats:
            unnorm_key = f"{unnorm_key}_no_noops"

    # Setup logging
    DATE_TIME = time.strftime("%Y_%m_%d-%H_%M_%S")
    os.makedirs(config.LOG_DIR, exist_ok=True)
    log_path = os.path.join(config.LOG_DIR, f"LITEN-{task_name}-{DATE_TIME}.txt")
    log_file = open(log_path, "w")

    log_file.write(f"LITEN Evaluation: {task_name}\n")
    log_file.write(f"Max Retries: {config.MAX_RETRIES}\n")
    log_file.write(f"Checkpoint: {checkpoint}\n")
    log_file.write("="*80 + "\n\n")

    # Metrics
    total_episodes = 0
    total_successes = 0
    total_retries = 0

    # Results storage for comparison
    results_per_task = []

    for task_id in range(num_tasks):
        task_info = task_suite_obj.get_task(task_id)
        init_states = task_suite_obj.get_task_init_states(task_id)

        # Create environment
        bddl_file = os.path.join(
            get_libero_path("bddl_files"),
            task_info.problem_folder,
            task_info.bddl_file
        )
        env = OffScreenRenderEnv(
            bddl_file_name=bddl_file,
            camera_heights=config.RESOLUTION,
            camera_widths=config.RESOLUTION,
        )
        env.seed(0)
        task_desc = task_info.language

        print(f"\n{'─'*80}")
        print(f"Task {task_id+1}/{num_tasks}: {task_desc}")
        print(f"{'─'*80}")
        log_file.write(f"\nTask {task_id+1}/{num_tasks}: {task_desc}\n")

        task_successes = 0

        for ep in range(config.NUM_TRIALS_PER_TASK):
            print(f"\n  Episode {ep+1}/{config.NUM_TRIALS_PER_TASK}:")
            log_file.write(f"\n  Episode {ep+1}:\n")

            # LITEN: Try with potential retries
            success = False
            attempt = 0
            failure_mode = None

            while not success and attempt < config.MAX_RETRIES:
                attempt += 1
                print(f"    Attempt {attempt}/{config.MAX_RETRIES}...", end=" ")

                success, steps, images, failure_mode = run_episode_liten(
                    env, model, processor, task_desc, init_states[ep],
                    max_steps, unnorm_key, attempt, failure_mode
                )

                if success:
                    print(f"Success! ({steps} steps)")
                    log_file.write(f"    Attempt {attempt}: SUCCESS ({steps} steps)\n")
                else:
                    print(f"✗ {failure_mode}")
                    log_file.write(f"    Attempt {attempt}: {failure_mode}\n")

            total_episodes += 1
            total_retries += (attempt - 1)
            if success:
                total_successes += 1
                task_successes += 1

            # Save final video
            if config.SAVE_VIDEOS and images:
                video_dir = f"{config.VIDEO_DIR}/{time.strftime('%Y_%m_%d')}"
                os.makedirs(video_dir, exist_ok=True)
                video_path = f"{video_dir}/LITEN-{DATE_TIME}-ep{total_episodes}-{'ok' if success else 'fail'}.mp4"
                writer = imageio.get_writer(video_path, fps=30)
                for img in images:
                    writer.append_data(img)
                writer.close()

        task_rate = task_successes / config.NUM_TRIALS_PER_TASK
        results_per_task.append({
            "task": task_desc,
            "success_rate": task_rate,
            "successes": task_successes
        })

        print(f"\n  Task Success: {task_rate:.0%} ({task_successes}/{config.NUM_TRIALS_PER_TASK})")
        log_file.write(f"\n  Task Success: {task_rate:.0%}\n")

        env.close()

    # Final summary
    overall_rate = total_successes / total_episodes
    avg_retries = total_retries / total_episodes

    print(f"\n{'='*80}")
    print(f"LITEN EVALUATION COMPLETE")
    print(f"{'='*80}")
    print(f"Total Episodes: {total_episodes}")
    print(f"Total Successes: {total_successes}")
    print(f"Overall Success Rate: {overall_rate:.1%}")
    print(f"Average Retries: {avg_retries:.2f}")
    print(f"{'='*80}\n")

    log_file.write(f"\n{'='*80}\n")
    log_file.write(f"Overall Success Rate: {overall_rate:.1%}\n")
    log_file.write(f"Average Retries: {avg_retries:.2f}\n")
    log_file.close()

    print(f"Log saved to: {log_path}")

    return {
        "task_suite": task_name,
        "success_rate": overall_rate,
        "avg_retries": avg_retries,
        "per_task": results_per_task
    }

# ============================================================================
# Run Experiments
# ============================================================================

# Run SPATIAL evaluation
print("\n" + "="*80)
print("Running LITEN on libero_spatial...")
print("="*80)
spatial_results = run_liten_evaluation("SPATIAL")

# Run OBJECT evaluation
print("\n" + "="*80)
print("Running LITEN on libero_object...")
print("="*80)
object_results = run_liten_evaluation("OBJECT")

# ============================================================================
# Comparison Summary
# ============================================================================

print("\n" + "="*80)
print("COMPARISON: BASELINE vs LITEN")
print("="*80)
print("\nBaseline Results (from your evaluation logs):")
print("  libero_spatial: 68.0% (17/25 episodes)")
print("  libero_object:  68.0% (17/25 episodes)")
print("\nLITEN Results:")
print(f"  libero_spatial: {spatial_results['success_rate']:.1%}")
print(f"    Avg retries: {spatial_results['avg_retries']:.2f}")
print(f"  libero_object:  {object_results['success_rate']:.1%}")
print(f"    Avg retries: {object_results['avg_retries']:.2f}")

improvement_spatial = spatial_results['success_rate'] - 0.68
improvement_object = object_results['success_rate'] - 0.68

print(f"\nImprovement:")
print(f"  libero_spatial: {improvement_spatial:+.1%}")
print(f"  libero_object:  {improvement_object:+.1%}")
print("="*80)
```
